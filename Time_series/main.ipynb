    Plotting time series using matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r'venv\ch2_co2_levels.csv')

#adding styles to plots
df = df.set_index('datestamp')
plt.style.use('dark_background')
print(plt.style.available)

#describing graphs with labels
ax = df.plot(color='blue', figsize=(12, 5), fontsize=12,linewidth=3,linestyle='--' )
ax.set_xlabel('Date', fontsize=16)
ax.set_ylabel('Y axis values', fontsize=16)
ax.set_title('time series graph', fontsize=16)
plt.show()
```

    Plotting subset of time series

```python
import pandas as pd
import matplotlib.pyplot as plt

df2 = pd.read_csv(r'venv/ch1_discoveries.csv')
print(df2.dtypes)
print('--------')
df2['date'] = pd.to_datetime(df2['date'])
print(df2.dtypes)

#TASK
df2 = df2.set_index('date')
df2.plot()
plt.style.use('classic')
df2_subset = df2['1860':'1888']
ax = df2_subset.plot(color='blue', fontsize=14,linestyle='--')
plt.axhline(y=4, color='g', linestyle='-')
plt.axvline('1875', color='r', linestyle='--')
plt.axvspan('1885', '1890', color='red', alpha=0.5)
ax.set_title('Time Series')
ax.set_xlabel('Date')
ax.set_ylabel('Number of discoveries')
plt.show()
```
